```python
import pandas as pd 
import numpy as np
file_data="Resources/purchase_data.csv"
purchase_data=pd.read_csv(file_data)
purchase_data.head()
```

```python
total_players=len(purchase_data["SN"].unique())
total_players
player_df=pd.DataFrame({"Total Players":[total_players]})
player_df
```

```python
num_uniqueitems=len(purchase_data["Item Name"].unique())
average_price=round(purchase_data["Price"].mean(),2)
total_purchases=len(purchase_data["Purchase ID"])
total_revenue=purchase_data["Price"].sum()
# print(num_uniqueitems)
# print(average_price)
# print(total_revenue)
#print(total_purchases)
summary_df=pd.DataFrame({"Number of Unique Items":[num_uniqueitems],
"Average Price":[average_price],
"Number of Purchases":[total_purchases],
"Total Revenue":[total_revenue]
})
summary_df["Average Price"] = summary_df["Average Price"].map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)

summary_df
```

```python
unique=purchase_data.drop_duplicates(subset=['SN'])
male_count=(unique['Gender'].values == 'Male').sum()
# male_count
female_count=(unique['Gender'].values == 'Female').sum()
# female_count
other_count=(unique['Gender'].values == 'Other / Non-Disclosed').sum()
# other_count
# male_count
male_percent=male_count/total_players*100
# male_percent
female_percent=female_count/total_players*100
# female_percent
other_percent=other_count/total_players*100
# other_percent
data={'Total Count':[male_count,female_count,other_count],
'Percentage of Players':[male_percent,female_percent,other_percent]}
df=pd.DataFrame(data,index=['Male',
'Female',
'Other/Non-Disclosed'])

df["Percentage of Players"] = df["Percentage of Players"].map("{:,.2f}%".format)

df
```

```python
gender_purchase_cnt = purchase_data.groupby(["Gender"])["Price"].count()
gender_purchase_cnt
```

```python
gender_purchase_cnt = purchase_data.groupby(["Gender"])["Price"].count()
gender_purchase_avg = purchase_data.groupby(["Gender"])["Price"].mean()
gender_purchase_total = purchase_data.groupby(["Gender"])["Price"].sum()

gener_unique_cnt = purchase_data.drop_duplicates(subset='SN')["Gender"].value_counts()

player_avg_purchase = gender_purchase_total / gener_unique_cnt

gender_purchase_df = pd.DataFrame({
    "Purchase Count": gender_purchase_cnt, 
    "Average Purchase Price": gender_purchase_avg, 
    "Total Purchase Value": gender_purchase_total, 
    "Avg Total Purchase per Person": player_avg_purchase
})

gender_purchase_df["Total Purchase Value"] = gender_purchase_df["Total Purchase Value"].map("${:,.2f}".format)
gender_purchase_df["Average Purchase Price"] = gender_purchase_df["Average Purchase Price"].map("${:,.2f}".format)
gender_purchase_df["Avg Total Purchase per Person"] = gender_purchase_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
gender_purchase_df
```

```python
age_bins = [0,9,14,19,24,29,34,39,100]
group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

players_Ages_unique_df = purchase_data.drop_duplicates(subset='SN')["Age"].to_frame()

players_Ages_unique_df["Age Ranges"] = pd.cut(players_Ages_unique_df["Age"], bins=age_bins, labels=group_names)

age_demographics_cnts = players_Ages_unique_df["Age Ranges"].value_counts()

SN_cnt = purchase_data.drop_duplicates(subset='SN')["SN"].count()

age_demographics_percents = age_demographics_cnts / SN_cnt

age_demographics_df = pd.DataFrame({
    "Total Count": age_demographics_cnts, 
    "Percentage of Players": age_demographics_percents
})

age_demographics_df['Percentage of Players'] = age_demographics_df['Percentage of Players'].map("{:,.2%}".format)
age_demographics_df = age_demographics_df.sort_index(ascending=True)

age_demographics_df
```

```python
# age_bins = [0,9,14,19,24,29,34,39,100]
# group_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
# players_Ages_unique_df = purchase_data.drop_duplicates(subset='SN')["Age"].to_frame()
# players_Ages_unique_df["Age Ranges"] = pd.cut(players_Ages_unique_df["Age"], bins=age_bins, labels=group_names)
# age_demographics_cnts = players_Ages_unique_df["Age Ranges"].value_counts()

age_data = purchase_data.loc[:,["Age", "Price"]]
age_purchase_sum = age_data.groupby(["Age"])["Price"].sum()
age_purchase_avg = age_data.groupby(["Age"])["Price"].mean()
age_purchase_cnt = age_data.groupby(["Age"])["Price"].count()

item_purchase_df = pd.DataFrame({
    "Purchase Count": age_purchase_cnt,
    "Item Price": age_purchase_avg,
    "Total Purchase Value": age_purchase_sum
})
item_purchase_df
```

```python
spender_data=purchase_data.loc[:,['SN','Price']]

spender_purchase_sum = spender_data.groupby(["SN"])["Price"].sum()
spender_purchase_avg = spender_data.groupby(["SN"])["Price"].mean()
spender_purchase_cnt = spender_data.groupby(["SN"])["Price"].count()

spender_purchase_df = pd.DataFrame({
    "Purchase Count": spender_purchase_cnt,
    "Item Price": spender_purchase_avg,
    "Total Purchase Value": spender_purchase_sum
})
spender_purchase_df["Item Price"] = spender_purchase_df["Item Price"].map("${:,.2f}".format)
spender_purchase_df["Purchase Count"] = spender_purchase_df["Purchase Count"].map("{:,}".format)
spender_purchase_df["Total Purchase Value"] = spender_purchase_df["Total Purchase Value"].map("${:,.2f}".format)
spender_purchase_df = spender_purchase_df.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]

spender_purchase_df.head(5)
```

```python
item_data_df = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

item_purchase_sum = item_data_df.groupby(["Item ID", "Item Name"])["Price"].sum()
item_purchase_avg = item_data_df.groupby(["Item ID", "Item Name"])["Price"].mean()
item_purchase_cnt = item_data_df.groupby(["Item ID", "Item Name"])["Price"].count()

item_purchase_df = pd.DataFrame({
    "Purchase Count": item_purchase_cnt,
    "Item Price": item_purchase_avg,
    "Total Purchase Value": item_purchase_sum
})

pop_item_sorted_df = item_purchase_df.sort_values("Purchase Count", ascending=False)

pop_item_sorted_df["Item Price"] = pop_item_sorted_df["Item Price"].map("${:,.2f}".format)
pop_item_sorted_df["Purchase Count"] = pop_item_sorted_df["Purchase Count"].map("{:,}".format)
pop_item_sorted_df["Total Purchase Value"] = pop_item_sorted_df["Total Purchase Value"].map("${:,.2f}".format)
pop_item_sorted_df = pop_item_sorted_df.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]

pop_item_sorted_df.head(5)
```

```python
popitem_data_df = purchase_data.loc[:,["Item ID","Item Name","Price"]]

item_purchase_sum = popitem_data_df.groupby(["Item ID", "Item Name"])["Price"].sum()
item_purchase_avg = popitem_data_df.groupby(["Item ID", "Item Name"])["Price"].mean()
item_purchase_cnt = popitem_data_df.groupby(["Item ID", "Item Name"])["Price"].count()

item_purchase_df = pd.DataFrame({
    "Purchase Count": item_purchase_cnt,
    "Item Price": item_purchase_avg,
    "Total Purchase Value": item_purchase_sum
})

pop_item_sorted_df = item_purchase_df.sort_values("Total Purchase Value", ascending=False)

pop_item_sorted_df["Item Price"] = pop_item_sorted_df["Item Price"].map("${:,.2f}".format)
pop_item_sorted_df["Purchase Count"] = pop_item_sorted_df["Purchase Count"].map("{:,}".format)
pop_item_sorted_df["Total Purchase Value"] = pop_item_sorted_df["Total Purchase Value"].map("${:,.2f}".format)
pop_item_sorted_df = pop_item_sorted_df.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]

pop_item_sorted_df.head(5)
```
